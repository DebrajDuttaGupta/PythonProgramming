```python
import requests
from tkinter import *
import tkinter.messagebox 

root = Tk()
root.title("Currency converter:")
root.geometry("700x400")
root.configure(background='#e6e5e5')

variable1 = StringVar(root)
variable2 = StringVar(root)
variable1.set("currency")
variable2.set("currency")

def RealTimeCurrencyConversion():
    from_currency = variable1.get()
    to_currency = variable2.get()
    amount = Amount1_field.get()

    if amount == "":
        tkinter.messagebox.showinfo("Error !!", "Amount Not Entered.\n Please enter a valid amount.")
        return

    if from_currency == "currency" or to_currency == "currency":
        tkinter.messagebox.showinfo("Error !!", "Currency Not Selected.\n Please select FROM and TO Currency from the menu.")
        return

    try:
        url = f"https://api.exchangerate-api.com/v4/latest/{from_currency}"
        response = requests.get(url)
        data = response.json()
        conversion_rate = data['rates'][to_currency]
        new_amount = float(amount) * conversion_rate
        Amount2_field.delete(0, END)
        Amount2_field.insert(0, str(round(new_amount, 4)))
    except Exception as e:
        tkinter.messagebox.showinfo("Error !!", f"Failed to fetch conversion rate.\n{str(e)}")

def clear_all():
    Amount1_field.delete(0, END)
    Amount2_field.delete(0, END)

CurrenyCode_list = ["USD", "INR", "CAD", "CNY", "DKK", "EUR"]

headlabel = Label(root, font=('lato black', 19, 'bold'), text='Currency converter : ', bg='#e6e5e5', fg='black')
headlabel.grid(row=0, column=0, sticky=W)

label1 = Label(root, font=('lato black', 15, 'bold'), text="\t Amount : ", bg="#e6e5e5", fg="black")
label1.grid(row=1, column=0, sticky=W)

label2 = Label(root, font=('lato black', 15, 'bold'), text="\t From Currency : ", bg="#e6e5e5", fg="black")
label2.grid(row=2, column=0, sticky=W)

label3 = Label(root, font=('lato black', 15, 'bold'), text="\t To Currency : ", bg="#e6e5e5", fg="black")
label3.grid(row=3, column=0, sticky=W)

label4 = Label(root, font=('lato black', 15, 'bold'), text="\t Converted Amount : ", bg="#e6e5e5", fg="black")
label4.grid(row=5, column=0, sticky=W)

Amount1_field = Entry(root)
Amount1_field.grid(row=1, column=1, ipadx=28, sticky=E)

Amount2_field = Entry(root)
Amount2_field.grid(row=5, column=1, ipadx=31, sticky=E)

FromCurrency_option = OptionMenu(root, variable1, *CurrenyCode_list)
ToCurrency_option = OptionMenu(root, variable2, *CurrenyCode_list)

FromCurrency_option.grid(row=2, column=1, ipadx=45, sticky=E)
ToCurrency_option.grid(row=3, column=1, ipadx=45, sticky=E)

button_convert = Button(root, font=('arial', 15, 'bold'), text=" Convert ", padx=2, pady=2, bg="lightblue", fg="white", command=RealTimeCurrencyConversion)
button_convert.grid(row=4, column=0)

button_clear = Button(root, font=('arial', 15, 'bold'), text=" Clear All ", padx=2, pady=2, bg="lightblue", fg="white", command=clear_all)
button_clear.grid(row=6, column=0)

root.mainloop()
```
